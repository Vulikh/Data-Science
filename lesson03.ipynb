```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
data = boston.data
data
```

```python
target = boston.target
target
```

```python
feature_names = boston.feature_names
```

```python
X = pd.DataFrame(data, columns=feature_names)
```

```python
X.head()
```

```python
y = pd.DataFrame(target, columns=['price'])
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score 
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000, max_depth = 12, random_state = 42)
```

```python
rf
```

```python
rf.fit(X_train, y_train.values[:, 0])
```

```python
y_rf_pred = rf.predict(X_test)
```

```python
r2_score(y_test, y_rf_pred)
```

Модель RandomForestRegressor в данном случае работает лучше, т.к. величина R2 ближе к единице

```python
rf.feature_importances_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.barh(X_train.columns, rf.feature_importances_.flatten())
plt.xlabel('Вес признака')
plt.ylabel('Признак')
```

Наибольшую важность показывают признаки LSTAT (% lower status of the population) и RM (average number of rooms per dwelling)
