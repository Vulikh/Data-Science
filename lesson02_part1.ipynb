```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color = 'green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
cc = pd.read_csv('creditcard.csv', sep = ',')
plt.style.use('fivethirtyeight')
cc.head()
```

```python
cc['Class'].value_counts().plot(kind = 'bar')
```

```python
cc['Class'].value_counts().plot(kind = 'bar', logy = True)
```

```python
cheats = cc[cc['Class'] == 1]['V1']
casual = cc[cc['Class'] == 0]['V1']
```

```python
fig, ax = plt.subplots()
plt.hist(casual, bins = 20, color = 'grey', alpha = 0.5, density = True, label = 'Class 0')
plt.hist(cheats, bins = 20, color = 'red', alpha = 0.5, density = True, label = 'Class 1')
plt.xlabel('Class')
legend = plt.legend(loc = 'upper left')
```
